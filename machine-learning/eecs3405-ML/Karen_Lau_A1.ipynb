### **Q1** (10 marks)  ###

Given two sets of $m$ vectors, $\mathbf{x}_i \in \mathbb{R}^n$  and $\mathbf{y}_i \in \mathbb{R}^n$ for all $i=1,2, \cdots, m$, verify that the summation $\sum_{i=1}^m  \mathbf{x}_i  \mathbf{y}_i^\intercal$ can be vectorized as the following matrix multiplication:
$$
\sum_{i=1}^m \mathbf{x}_i \mathbf{y}_i^\intercal = \mathbf{X}  \mathbf{Y}^\intercal,
$$
where $\mathbf{X} = \big[ \mathbf{x}_1 \, \mathbf{x}_2 \, \cdots \, \mathbf{x}_m \big] \in \mathbb{R}^{n\times m}$ and $\mathbf{Y} = \big[ \mathbf{y}_1 \, \mathbf{y}_2 \, \cdots \, \mathbf{y}_m \big] \in \mathbb{R}^{n\times m}$.

**Answer:**

We have
$\mathbf{X} = \big[ \mathbf{x}_1 \, \mathbf{x}_2 \, \cdots \, \mathbf{x}_m \big]$  and $\mathbf{Y} = \big[ \mathbf{y}_1 \, \mathbf{y}_2 \, \cdots \, \mathbf{y}_m \big]$, where each $\mathbf{x_i}$ and $\mathbf{y_i}$ are column vectors.

If we compute $\mathbf{X}  \mathbf{Y}^\intercal$, we obtain an $n$ x $n$ matrix, such that each (p-q)-th entry:

$$(\mathbf{X} \mathbf{Y}^\intercal)_{pq} = \sum_{k=1}^m \mathbf{X}_{pk} \mathbf{Y}_{qk}$$

Each $\mathbf{X}_{pk}$ is the p-th component of the k-th column $\mathbf{x}_k$ and $\mathbf{Y}_{qk}$ is the q-th component of the k-th column $\mathbf{y}_k$.

Thus $$(\mathbf{X} \mathbf{Y}^\intercal)_{pq} = \sum_{k=1}^m (\mathbf{x}_{k})_p (\mathbf{y}_{k})_q$$

But this is also exactly equivalent to
$\sum_{i=1}^m \mathbf{x}_i \mathbf{y}_i^\intercal$. Since:

$$
(\sum_{i=1}^m \mathbf{x}_i \mathbf{y}_i^\intercal)_{pq} = \sum_{i=1}^m (\mathbf{x}_i \mathbf{y}_i^\intercal)_{pq} = \sum_{i=1}^m (\mathbf{x}_i)_p (\mathbf{y}_i)_{q}
$$

Since all of the entries/elements match, we can conclude:
$$
\sum_{i=1}^m \mathbf{x}_i \mathbf{y}_i^\intercal = \mathbf{X}  \mathbf{Y}^\intercal,
$$


### **Q2** (10 marks)  ###

Given $\mathbf{x} \in \mathbb{R}^n$, $\mathbf{z} \in \mathbb{R}^m$, and
$\mathbf{A} \in \mathbb{R}^{m \times n}$ ($m < n$), compute the following two partial derivatives:

1.   $\frac{\partial}{\partial \mathbf{z}} \, \Vert  \mathbf{A} \mathbf{x} - \mathbf{z} \Vert^2$

2.   $\frac{\partial}{\partial \mathbf{x}} \, \Vert  \mathbf{A} \mathbf{x} - \mathbf{z} \Vert^2$



**ANSWER:**

We know that:
$$
\begin{aligned}
\Vert  \mathbf{A} \mathbf{x} - \mathbf{z} \Vert^2 &= (\mathbf{A} \mathbf{x} - \mathbf{z})^\intercal(\mathbf{A} \mathbf{x} - \mathbf{z}) \\
          &=  ((\mathbf{A} \mathbf{x})^\intercal - \mathbf{z}^\intercal)(\mathbf{A} \mathbf{x} - \mathbf{z}) \\
          &= (\mathbf{A} \mathbf{x})^\intercal(\mathbf{A} \mathbf{x}) - 2\mathbf{z}^\intercal(\mathbf{A} \mathbf{x}) + \mathbf{z}^\intercal\mathbf{z}
\end{aligned}
$$

We can now find the partial derivatives.
1. $$
\begin{aligned}
 \frac{\partial}{\partial \mathbf{z}} \, \Vert  \mathbf{A} \mathbf{x} - \mathbf{z} \Vert^2 &=  \frac{\partial}{\partial \mathbf{z}} ((\mathbf{A} \mathbf{x})^\intercal(\mathbf{A} \mathbf{x}) - 2\mathbf{z}^\intercal(\mathbf{A} \mathbf{x}) + \mathbf{z}^\intercal\mathbf{z}) \\
 &= -2(\mathbf{A} \mathbf{x}) + 2\mathbf{z}\\
        &= 2(\mathbf{z} - \mathbf{A}\mathbf{x})
\end{aligned}
$$


2.  $$
\begin{aligned}
 \frac{\partial}{\partial \mathbf{x}} \, \Vert  \mathbf{A} \mathbf{x} - \mathbf{z} \Vert^2 &=  \frac{\partial}{\partial \mathbf{x}} ((\mathbf{A} \mathbf{x})^\intercal(\mathbf{A} \mathbf{x}) - 2\mathbf{z}^\intercal(\mathbf{A} \mathbf{x}) + \mathbf{z}^\intercal\mathbf{z})\\
 &= 2\mathbf{A}^\intercal\mathbf{A}\mathbf{x} - 2\mathbf{A}^\intercal\mathbf{z}\\
 &= 2\mathbf{A}^\intercal(\mathbf{A}\mathbf{x}-\mathbf{z})
\end{aligned}
$$

**Part 3.1 (10 marks):** Derive a gradient decent algorithm to solve the following quadratic optimization problem:

$$
    \min_{\mathbf{x}} \;\; \frac12 \mathbf{x}^\intercal \mathbf{Q} \mathbf{x} - \mathbf{1}^\intercal \mathbf{x}
$$

where $\mathbf{x} \in \mathbb{R}^n$ is free variables, $\mathbf{1}$ is a constant $n$-dimensional vector whose elements are all 1,  and $\mathbf{Q} \in \mathbb{R}^{n \times n}$ is given as a constant definite positive matrix ($\mathbf{Q} \succ 0$).

**Answer:**

First, we must find the gradient of the function:

$$
   f(x) =  \frac12 \mathbf{x}^\intercal \mathbf{Q} \mathbf{x} - \mathbf{1}^\intercal \mathbf{x}
$$
$$\nabla f(x) = \frac{\partial}{\partial \mathbf{x}} f(x) = \mathbf{Q} \mathbf{x} - \mathbf{1} $$

Then, the gradient descent algorithm is:

$$x_{k+1} = x_k - n_n(\mathbf{Q}\mathbf{x}_k - \mathbf{1})$$



**Part 3.2 (10 marks):**  Use `numpy` to write a gradient descent code to solve the above quadratic optimization problem.  Print the solution $\mathbf{x}^*$ for the following $\mathbf{Q}$ matrix.

```python
import numpy as np

Q = np.array([[13.57599411,-7.77916099,4.45073215,0.18157029,2.55400279,-1.06271637,-0.14237939,-4.73200647],
 [-7.77916099,9.16482133,1.349321,0.4697293,-3.29959602,3.41477056,-4.66924506,1.74397411],
 [4.45073215,1.349321,11.66150687,4.7431563,4.6278791,0.61248024,-2.52679826,-4.94462079],
 [0.18157029,0.4697293,4.7431563,13.50647727,8.16622632,2.77908368,-2.36985429,-7.53962758],
 [2.55400279,-3.29959602,4.6278791,8.16622632,12.2258951,-1.84304299,1.04977205,-8.24939242],
 [-1.06271637,3.41477056,0.61248024,2.77908368,-1.84304299,6.67184174,-4.60482886,0.28920379],
 [-0.14237939,-4.66924506,-2.52679826,-2.36985429,1.04977205,-4.60482886,13.16925778,3.41548824],
 [-4.73200647,1.74397411,-4.94462079, -7.53962758,-8.24939242,0.28920379,3.41548824,8.72679051]])

print(Q.shape)
```

```python
np.random.seed(0)
x = np.random.rand(Q.shape[0])     #Choosing a starting point
step = 0.9 / np.max(np.linalg.eigvals(Q))     #Step size
tolerance = 1e-8                        #tolerance for convergence
max_iter = 10000          #maximum number of iterations
n=0           #number iterations completed

while n < max_iter:
    grad = Q @ x - np.ones(Q.shape[0])   #computes gradient
    x_new = x - step * grad               #update

    if np.linalg.norm(x_new - x) < tolerance:
        break

    x = x_new                   #adjust
    n += 1

print("Converged after", n, "iterations")
print("x* =", x)
```

```python
#Check for convergence
x_true = np.linalg.solve(Q, np.ones(Q.shape[0]))
print(np.allclose(x, x_true, atol=1e-4))
```

**Part 4.1 (10 marks):** Derive the procedure to compute the distance of a point $\mathbf{x}_0 \in \mathbb{R}^n$ to
an elliptic surface $\mathbf{x}^\intercal \mathbf{A} \mathbf{x} = 1$, where $\mathbf{A} \in \mathbb{R}^{n \times n}$ is symmetric and positive definite ($\mathbf{A} \succ 0$).

*Hints: Use the method of Lagrange multipliers to derive the formula, and may need to use a numerical procedure (e.g. grid search) to find out the unknown multiplier(s) if no closed-form solution exists.*



**Answer:**

We want to compute the distance of a point $\mathbf{x_o}$ to the elliptic surface  $\mathbf{x}^\intercal \mathbf{A} \mathbf{x} = 1$.

Therefore, we are trying to minimize $d^2 = \Vert \mathbf{x} - \mathbf{x_o}\Vert ^2$ subject to
$\mathbf{x}^\intercal \mathbf{A} \mathbf{x} = 1$

In order to do this, we can introduce the Lagrange multiplier $\lambda$, to get:

$$
\begin{aligned}
L(x,\lambda) &= \Vert \mathbf{x} - \mathbf{x_o}\Vert ^2 + \lambda(\mathbf{x}^\intercal \mathbf{A} \mathbf{x} -1)\\
&= (\mathbf{x} - \mathbf{x_o}\intercal (\mathbf{x}-\mathbf{x_o}) + \lambda(\mathbf{x}^\intercal \mathbf{A} \mathbf{x} -1))
\end{aligned}
$$

Then we want to find when $\frac{\partial \mathbf{L(x,λ)}}{\partial \mathbf{x}} = 0$

Which is:
$$
\begin{aligned}
 \frac{\partial \mathbf{L(x,λ)}}{\partial \mathbf{z}} &= 2(\mathbf{x} - \mathbf{x_o}) + 2\lambda\mathbf{Ax} \\
0 &= \mathbf{x} - \mathbf{x_o} + \lambda\mathbf{Ax} \\
\mathbf{x_o}&= \mathbf{x}(I + \lambda\mathbf{A})\\
\mathbf{x^*} &= (I + \lambda\mathbf{A})^{-1} \mathbf{x_o}
\end{aligned}
$$

Then we must solve for $\lambda$:

We have that $\mathbf{x}^\intercal \mathbf{A} \mathbf{x} = 1$, we can substitute $\mathbf{x^*}$, such that we obtain:

 $((I + \lambda\mathbf{A})^{-1} \mathbf{x_o})^\intercal \mathbf{A} (I + \lambda\mathbf{A})^{-1} \mathbf{x_o} - 1 = 0$




**Part 4.2 (10 marks):** Use `numpy` to implement a grid search to compute the above distance.
Use the following given $\mathbf{x}_0$ and $\mathbf{A}$ as an example to print out the distance.

```python
import numpy as np

A = np.array([[10.82133645,3.92407517,0.28991746,-2.03411727,4.55376106],
 [3.92407517,8.55336068,-0.21915151,0.42212004,2.08124307],
 [0.28991746,-0.21915151,5.74622299,-0.800496,0.63366946],
 [-2.03411727,0.42212004,-0.800496,3.91232017,-0.44196245],
 [4.55376106,2.08124307,0.63366946,-0.44196245,3.83040884]])

x0 = np.array([-0.08111786,1.68122205,0.84108852,1.74916213,-0.11266981])

print(A.shape)
print(x0.shape)
```

```python
# Grid of positive lambda values
lambdas = np.linspace(0, 10, 10000)

best_lambda = None  #initialize lambda
best_error = np.inf

for lam in lambdas:   #loops over all lambdas
    M = np.linalg.inv(np.eye(A.shape[0]) + lam * A)
    x_star = M @ x0
    error = abs(x_star.T @ A @ x_star - 1)
    if error < best_error:
        best_error = error
        best_lambda = lam

# Compute x* and distance
x_star_best = np.linalg.inv(np.eye(A.shape[0]) + best_lambda * A) @ x0
distance = np.linalg.norm(x_star_best - x0)

print("Best λ:", best_lambda)
print("Constraint value:", x_star_best.T @ A @ x_star_best)
#checks if estimation is good

print("Distance:", distance)
```
