### Question 1 -- Data science terminology (5 Marks)
Read [this news article](https://www.wired.com/story/labgenius-antibody-factory-machine-learning/). Make up an example of a data science process which involves the model discussed in this article.



Question 1 Response:

The data science process is a 4 step cycle: capture, maintain, analysis and communicate. Once the cycle completes one round, it will loop and repeat itself.

Capturing involves collecting and receiving data. Maintaining involves storing the data to be easily accessed and analyzed for the next step. Analysis involves manipulating and transforming the data into useful information to answer questions.
Finally, communication is simply reporting any new and relevant information from the analysis of the data that was collected. 

The article discusses a machine-learning model which designs particular antibodies for fighting specific diseases. This model requires little human input and will continuously feed into the data science process as it will feed the information it has found back into the model. 

<b> An example of when we could use this model could be the following:

It was recently observed that there has been a spike in cases of a new COVID-19 variant infection. This variant is immune to the existing vaccinations and evade the antibodies produced by them. Therefore, a new antibody must be developed to prevent the further spread of the disease, this can be completed with the model discussed in the article. 

1. Capture:
    - Collecting data on the increased infection rates
    - Collect information about the new COVID-19 variant - structure
    - Collect information about how the previous antibody interacts with
      the new variant structure

2. Maintain
    - Store the collected data in a database. 
   
3. Analysis
    - Input data into the model
    - Model will begin testing antibody structures
    - Determine which antibody structures are the best 
   
4. Communication
    -  Present results of best antibody structures
  
If necessary, that is, if further modification of the antibody structures are required, the information found from part 3/4 will be fed back into part 1 of the next cycle. 
The results of this model can then be used to manufacture new vaccinations for the novel COVID-19 Variant.
