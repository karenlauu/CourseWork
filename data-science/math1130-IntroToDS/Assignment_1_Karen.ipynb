Control maps can be found in the dataset by rows where the `control_perecent` is not 0 for at least one team.
````python
#Question 3d, part 1 - data prep/win rate

#Counts the number of wins/losses for each team
wins = subset_df['map_winner'].value_counts()
losses = subset_df['map_loser'].value_counts()

#creates a dataframe with the wins and losses in each column, rows as teams 
#fills in empty data with 0 
rate_df = pd.DataFrame(
    {'wins': wins, 'losses':losses}).fillna(0)

#Finds the total number of matches each team played into a new column
rate_df['total_matches'] = rate_df['wins'] + rate_df['losses']

#Calculates each team's win rate
rate_df['win_rate'] = rate_df['wins']/rate_df['total_matches']

#names the index column as teams
rate_df.index.name = 'teams'

#converts index into a column
rate_df = rate_df.reset_index()

#adding the countries into the df
rate_df['country'] = rate_df['teams'].apply(lambda x: country[x])

#ordering the data from lowest to highest
rate_df = rate_df.sort_values('win_rate')
````

````python
#Question 5c
#Assign each team a rank according to their 2022 control map win rate, with 1 being the highest win rate on control maps in 2022 and 
#20 being the lowest win rate on control maps in 2022. Print each team name with its rank beside it.

#accessing data for maps and wins/losses per team
winsmaps_2022 = subset_df[subset_df['year'] == 2022].groupby(['control_round_name', 'map_winner'])['map_winner'].count()
winsmaps_2022.index.names = ['control map', 'team']
winsmaps_2022 = winsmaps_2022.reset_index()
winsmaps_2022

lossmaps_2022 = subset_df[subset_df['year'] == 2022].groupby(['control_round_name', 'map_loser'])['map_loser'].count()
lossmaps_2022.index.names = ['control map', 'team']
lossmaps_2022 = lossmaps_2022.reset_index()

#joins the wins and losses dataframes together
maps_2022 = pd.merge(
    winsmaps_2022,
    lossmaps_2022,
    on = ['control map','team'],
    how = 'outer'
)

#fills in 0s for any team that has 0 wins or 0 losses
maps_2022['map_winner'] = maps_2022['map_winner'].fillna(0)
maps_2022['map_loser'] = maps_2022['map_loser'].fillna(0)

#Calculates the win rate for each team for each map 
maps_2022['total matches'] = maps_2022['map_winner'] + maps_2022['map_loser']
maps_2022['win rate'] =  maps_2022['map_winner'] / maps_2022['total matches']

#Adds rank for the teams for each map
maps_2022['rank'] = maps_2022.groupby('control map')['win rate'].rank(ascending=False, method='first')
maps_2022 = maps_2022.sort_values(['control map', 'rank'])

#Printing the team ranks
for map, group in maps_2022.groupby('control map'):
    print("***** Control Map: " + map + " *****")
    
    for i in range(len(group)):
        team = group.iloc[i]['team']
        rank = int(group.iloc[i]['rank']
              )
        print("Team: " + team + ", Rank: " + str(rank))
    print('-'* 15) 

##order by rank
````

````python
#Question 5d, cont'd - mean of other teams

#Toronto Defiant may not have played all teams at a map > Creates a data frame to check the teams they played for each map
#Filters dataframe to 2022, and if Toronto Defiants played 
torontomatches_2022 = subset_df[(subset_df['year'] == 2022) & 
    ((subset_df['map_winner'] == 'Toronto Defiant') | (subset_df['map_loser'] == 'Toronto Defiant'))]

#Filters df to only these three columns
torontomatches_2022 = torontomatches_2022[['control_round_name', 'map_winner', 'map_loser']]
torontomatches_2022.sort_values('control_round_name')
torontomatches_2022['control_round_name'] = torontomatches_2022['control_round_name'].astype(str)

#creating a dictionary with list of opponent names for each map
toronto_opponents = {}

toronto_maps = top3_toronto + bottom3_toronto

#Loops through each map to collect opponent names and puts them into a dictionary
for map in toronto_maps:
    games = torontomatches_2022.loc[torontomatches_2022['control_round_name'] == map]

    #Opponents when toronto is the map winner
    if any(games['map_winner'] == "Toronto Defiant"):
        opponents = games['map_loser'][games['map_winner']=='Toronto Defiant']
        toronto_opponents[map] = list(set(opponents.tolist()))
    
    #Opponents when toronto is the map loser
    if any(games['map_loser'] == "Toronto Defiant"):
        opponents = games['map_winner'][games['map_loser']=='Toronto Defiant']
        if map in toronto_opponents:
            toronto_opponents[map].extend(opponents.tolist())
            toronto_opponents[map] = list(set(toronto_opponents[map]))
        else:
            toronto_opponents[map] = list(set(opponents.tolist()))


#Using the dictionary and df with ranks (from c) to get the mean rank for the opponents on each map
#Filter the dataframe to only the maps, teams and ranks 
opponent_ranks = maps_2022[['control map', 'team', 'rank']]

average_opponent_rank = {}

for map, opponents in toronto_opponents.items():
    map_ranks = opponent_ranks[(opponent_ranks['control map'] == map) & (opponent_ranks['team'].isin(opponents))]
    average_rank = map_ranks['rank'].mean()
    average_opponent_rank[map] = average_rank

for map, rank in average_opponent_rank.items():
    rounded_avg_rank = round(rank,2)
    print('Map: ' + map + ", Average Opponent Rank: " + str(rounded_avg_rank)) 
````
